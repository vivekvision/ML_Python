```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# header is not avaible and also not required 
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
```

```python
dataset.head(2)
```

Every row of data repesents a purchase transaction by customer and all the items in that transaction as different columns


Find associations - find rules like "if customer buys this product customer will also buy that product"

```python
# Apriori model expect data in this format : list of list
dataset = dataset.astype('str')
transactions = dataset.values.tolist()
```

# Train Apriori Model

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2,\
                min_lift = 3, min_length = 2, max_length = 2)
```

min_support hyper param - is based on intuition that item appearing in minimum 3 transaction should be considered 
Its weekly data so weekly min support parma : (3*7)/(total transaction)

minimum confidence by experiment 0.2
minimum lift - good lift is at lest 3

minimum length - minimum number of elements required in the rule 
maximum length - maximum number of elements required in the rule 


```python
results = list(rules)
```

```python
results
```

items_base  means frist item
items_add - if first item or base item is purchase then high chance that items add will be purchased

support confidence & lift for each run has usual rule of Apriori 


# Format the reusult into dataframe

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), \
                                  columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
